```python

#Final Project
import os
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf

# Web scaping an etf ticker's data from an excel file within a download link. 
stocks = pd.read_excel('https://www.ssga.com/us/en/intermediary/etfs/library-content/products/fund-data/etfs/us/holdings-daily-us-en-spy.xlsx',
                        header=4).Ticker.dropna().to_list()

stocks = [ticker.upper() for ticker in stocks] 

# Gathers asset's data from past 5 years (1825 days) to today 
start_date = dt.datetime.today() - pd.Timedelta(1825, "d")
end_date = dt.datetime.today()

#Downloads stock data
data =yf.download (start=start_date,end=end_date, tickers= stocks)['Adj Close']
data= pd.DataFrame(data).dropna(how='all', axis=1) 

data2 =yf.download (start=start_date,end=end_date, tickers= stocks )
data2= pd.DataFrame(data2).dropna(how='all', axis=1)

data3 =yf.download (start=start_date,end=end_date, tickers= stocks )['Volume'] #for volume trendline
data3= pd.DataFrame(data).dropna(how='all', axis=1) 
```

52 weeks trendline slopes

```python
slopes=pd.DataFrame(index=data.columns)

#for trendline: use np.polyfit

def trendline (data):
    for ind in range (1,21):
        slopes[ind]=np.polyfit(range(252),data[-252 - ind : - ind], 1)[0] #collect this into empty DF, slopes . using range 252 days but step back once each time to get 20 day avg.

    return slopes.mean(axis=1) #returns avg slope of last 20 days by column(thetickers)

trendline(data=data.pct_change())
```

Percent above 260 day low

```python
#Low from 260 days
price_low = data2['Low'][-260:][-20:]
price_low
#current and lowest price
current_price= price_low.iloc[-1:]
lowest_price= price_low.min()

#percentage above 260 days
precentabove_lows =((current_price - lowest_price) / lowest_price) * 100
precentabove_lows
```

4/52 week price oscillator

```python
#52 weeks
fiftytwo_weekclose= data2['Close'][-252:]
fiftytwo_weekclose=fiftytwo_weekclose.ewm(span=252, min_periods=20).mean()[-20:].dropna()

#4 weeks
fourweek_close= data2['Close'][-20:]
fourweek_close=fourweek_close.ewm(span=20, min_periods=20).mean().dropna()

#price oscillator
price_oscillator =((fourweek_close-fiftytwo_weekclose)/fiftytwo_weekclose)*100
price_oscillator.dropna()
```

51 week volume price trend

```python
vol = data2['Volume'][-240:]
price_close= data2['Close'][-240:]

# Volume Price Trend 

VPT = [0]  # Initial VPT 

for i in range(1, 21):
    volume = vol.iloc[i]
    close_price = price_close.iloc[i]
    previous_close_price = price_close.iloc[i - 1]
    previousvpt = VPT[-1]  # Gets previous VPT value
    vpt = (volume * (close_price - previous_close_price / previous_close_price))+ previousvpt 
    
vpt
```

Z-scores

```python
import scipy
import scipy.stats as stats

data = data.pct_change()

zscore = scipy.stats.zscore(data, axis=0, ddof=0, nan_policy='omit')
zscore
```

10 Long and Short Baskets

```python
mean_zscores = zscore.mean().sort_values(ascending=False)

long_assets = mean_zscores.head(10).index.tolist()  # Adjusts to basket size 
short_assets = mean_zscores.tail(10).index.tolist()
print("Long Basket:",long_assets)
print("Short Basket:",short_assets)
```
