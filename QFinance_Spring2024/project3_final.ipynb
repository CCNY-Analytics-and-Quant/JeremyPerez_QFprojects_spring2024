```python
#Final Project
import os
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf


# select 100 stocks from S&P500 index (
stocks = [
    'MSFT', 'AAPL', 'NVDA', 'AMZN', 'META', 'GOOGL', 'C', 'LLY', 'AVGO', 'JPM',
    'XOM', 'TSLA', 'V', 'UNH', 'MA', 'PG', 'JNJ', 'HD', 'COST', 'MRK',
    'ABBV', 'CRM', 'CVX', 'NFLX', 'AMD', 'WMT', 'BAC', 'PEP', 'KO', 'TMO',
    'LIN', 'ADBE', 'DIS', 'WFC', 'ACN', 'CSCO', 'MCD', 'ORCL', 'QCOM','ABT',
    'CAT', 'INTU', 'AMAT', 'GE', 'VZ', 'IBM', 'DHR', 'NOW', 'CMCSA', 'UBER',
    'COP', 'TXN', 'INTC', 'PFE', 'AMGN', 'UNP', 'PM', 'ISRG','MU', 'SPGI',
    'LOW', 'RTX', 'NEE', 'HON', 'ETN', 'GS', 'LRCX', 'AXP', 'BKNG','PGR',
    'ELV', 'T', 'SYK', 'NKE', 'PLD', 'MS', 'TJX', 'BLK', 'MDT', 'UPS',
    'SCHW', 'DE', 'CI', 'VRTX', 'ADP', 'BSX', 'CB', 'BMY', 'MMC','BA', 
    'LMT', 'SBUX', 'ADI', 'REGN', 'KLAC', 'FI', 'MDLZ', 'BX', 'PANW','CVS',
]
stocks = [ticker.upper() for ticker in stocks] # alphabitize the dataset

# Gathers asset's data from past 5 years (1825 days) to today 
start_date = dt.datetime.today() - pd.Timedelta(1825, "d")
end_date = dt.datetime.today()

#Downloads stock data
stocks_data = yf.download (start=start_date,end=end_date, tickers= stocks )['Adj Close'] #slices Adj Close columns
df_stocks=pd.DataFrame(stocks_data)

df_stocks=df_stocks[-252:]  # 252 trading days in a year


```

```python
daily_returns=df_stocks.pct_change().dropna()
daily_returns_df=pd.DataFrame(daily_returns)
daily_returns_df['returns']=daily_returns.mean(axis=1) # avg of the returns of stocks row-wise 

daily_returns_df['direction'] = np.sign(daily_returns_df['returns']).astype(int) # pos/neg sign of the returns indicates direction of slope

# Selecting desired columns
daily_returns_df = daily_returns_df.loc[:, [ 'returns', 'direction']]
daily_returns_df
```

```python
#20 day lag
lags=20
def create_lags(daily_returns_df): 
    global cols
    cols = []
    for lag in range(1, lags + 1):
        col = 'lag_{}'.format(lag)
        daily_returns_df[col] = daily_returns_df['returns'].shift(lag)
        cols.append(col)
create_lags(daily_returns_df)

daily_returns_df.head()
```

```python
daily_returns_df.dropna(inplace=True)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
daily_returns_df['pos_ols_1'] = model.fit(daily_returns_df[cols], daily_returns_df['returns']).predict(daily_returns_df[cols])

daily_returns_df['pos_ols_2'] = model.fit(daily_returns_df[cols], daily_returns_df['direction']).predict(daily_returns_df[cols])

daily_returns_df[['pos_ols_1', 'pos_ols_2']].head()
```

```python
daily_returns_df[['pos_ols_1', 'pos_ols_2']] = np.where( daily_returns_df[['pos_ols_1', 'pos_ols_2']] > 0, 1, -1)
daily_returns_df['pos_ols_1'].value_counts()
daily_returns_df['pos_ols_2'].value_counts()
(daily_returns_df['pos_ols_1'].diff() != 0).sum()
(daily_returns_df['pos_ols_2'].diff() != 0).sum()
daily_returns_df['strat_ols_1'] = daily_returns_df['pos_ols_1'] * daily_returns_df['returns']
daily_returns_df['strat_ols_2'] = daily_returns_df['pos_ols_2'] * daily_returns_df['returns']
daily_returns_df[['returns', 'strat_ols_1', 'strat_ols_2']].sum().apply(np.exp)
(daily_returns_df['direction'] == daily_returns_df['pos_ols_1']).value_counts()
(daily_returns_df['direction'] == daily_returns_df['pos_ols_2']).value_counts()
```
